```python
# Marking the success/failed launches on the map
# class = 1 (success) --> green marker
# class = 0 (failure) --> blue marker

#creating a marker cluster object
marker_cluster = MarkerCluster()
def marker_color(val):
    if val == 1:
        return 'green'
    else:
        return 'blue'
df['color'] = df['class'].apply(lambda x : marker_color(x))

for idx, row in df.iterrows():
    marker = folium.Marker(location = [row['Lat'], row['Long']], 
                           icon = folium.Icon(color = 'white', icon_color = row['color']))
    marker_cluster.add_child(marker)

launch_map.add_child(marker_cluster)
```

```python
# calculating the distances between a launch site to its proximities
formatter = "function(num) {return L.Util.formatNum(num, 5);};" # 5indicates the decimals to be included

# creating a mouse position object
mouse_position = MousePosition(
        position = 'topright',
        separator = ' Long:',
        empty_string = 'NaN',
        lng_first=False, #--> longitude first  = False
        num_digits = 20,
        prefix='Cordinates | Lat:',
        lat_formatter=formatter,
        lng_formatter=formatter
)

launch_map.add_child(mouse_position)
launch_map
```

```python
# marking distances from highways, railways and coasts
from math import sin, cos, sqrt, atan2, radians

#cords_coast_1 : Lat = 28.56311, long = -80.56685 (near ccafs_slc)
#cords_coast_2 : Lat = 34.63603, Long = -120.62488 (near vafb)

#ccafs_lc = [28.562302, -80.577356]
#ccafs_slc = [28.563197, -80.576820]
#ksc_lc = [28.573255, -80.646895]
#vafb = [34.632834, -120.610745]


def calculate_distance(lat1, lon1, lat2, lon2):
    Rad = 6373 #earth radius in km
    
    lat1 = radians(lat1) # converting all cordinates to radians
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = Rad * c
    return distance
```

```python
dist_coast_1 = calculate_distance(28.562302, -80.577356, 28.56349, -80.56785)
dist_coast_2 = calculate_distance(28.563197, -80.576820, 28.56349, -80.56785)
dist_coast_3 = calculate_distance(28.573255, -80.646895, 28.56370, -80.56722 )
dist_coast_4 = calculate_distance(34.632834, -120.610745, 34.63603, -120.62488 )



print (f'Nearest coastline from CCAFS_LC is {dist_coast_1:.2f} km')
print (f'Nearest coastline from CCAFS_SLC is {dist_coast_2:.2f} km')
print (f'Nearest coastline from KSC_LC is {dist_coast_3:.2f} km')
print (f'Nearest coastline from VAFB is {dist_coast_4:.2f} km')
```
