### Reading the dataset from the last part

```python
df = pd.read_csv(r'dataset_part_1.csv')
df.head()
```

```python
#number of launches on each site
df['LaunchSite'].value_counts()
```

```python
df[df['LaunchSite'] == 'Kwajalein Atoll']
```

```python
# Calculating the occurence of each orbit
df['Orbit'].value_counts()
```

```python
# Calculting the number and occurence of mission outcome of the orbits
landing_outcomes = df['Outcome'].value_counts()
for i,outcome in enumerate(landing_outcomes.keys()):
    print(i,outcome)
    
```

```python
# getting the ouctomes where booster did not land successfully from the corresponding indexes, 2,3,5,6,7
bad_outcomes=set(landing_outcomes.keys()[[2,3,5,6,7]])
bad_outcomes
```

```python
# landing_class = 0 for bad outcome and 1 for not
landing_class = []
for x in df['Outcome']:
    if x in bad_outcomes:
        landing_class.append(int(0))
    else:
        landing_class.append(int(1))
landing_class
```

```python
df['Class'] = landing_class
df[['Class']].head(8)
```

```python
# calculating the success rate
df['Class'].mean()
```
